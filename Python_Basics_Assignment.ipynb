```python
# 3.Compare and contrast mutable and immutable objects in python with examples.

# Ans:

# In Python, objects can be classified as mutable and immutable based on whether their value can be changed after they are created.

# Mutable Objects : 

#Values/contents of mutable objects can be modified after they are created keeping the memory location as it is.
# Examples: Lists, dictionaries, sets

list1 = [1, 2, 3]
list1[0] = 10  # Modify the first element
print(list1)  # Output: [10, 2, 3]

# Here, the sample list list1 is mutable as we can change the vales after creation.
# Mutable objects are useful for collections that need to be updated frequently.

# Immutable Objects :

# Immutable objects cannot be changed after they are created. Rather we have to create a new object with a new memory location.
# Examples: Integers, floats, strings, tuples, frozen sets

my_str = "Hello"
my_str[0] = "h"  # This will raise an error

# Immutable objects are useful for storing a fixed amount of data that need not to be changed further.
```

```python
# 4. Discuss different types of operators in Python and provide the examples how they are used.

#Ans:

# In Python, operators are special symbols or keywords that perform operations on variables and values.
# Python provides various types of operators, each designed for specific operations. These include arithmetic, comparison, logical, bitwise, assignment, identity, membership etc.

# 1.Arithmatic operator

# Addition (+), Subtraction (-), Multiplication (*), Division (/), Modulus (%), Floor Division (//) and Exponention (**) come under Arithmatic operator.
# Ex:

print (3+4)
print (9-4)
print (6*3)
print (5 / 2)
print (5 % 2)
print (5 // 2)
print (2 ** 3)

# 2.Comparison operator

# These operators compare values and return boolean values like True or False.
# Equal to (==), Not equal to (!=), Greater than (>), Less than (<), Greater than or equal to (>=), Less than or equal to (<=) comes under comparision Operator.
# Ex:

print (3 == 5)
print (3 == 3)
print (3 > 5)
print (3 < 5)
print (3 >= 5)
print (5 >= 3)

# 3.Logical operator

# Logical operators are used to combine conditional statements.
# Ex:

a = True
b = False
print(a and b)  # Output: False
print(a or b)   # Output: True
print(not a)    # Output: False

# 4.Assignment operator

# Assignment operator is used to assign a value to a perticular variable.
# ==, +=, -=, *=, /=, %=, **=, //= are used as assignment operators in python.

x = 100

print (x == 100)

x += 5
print (x)

x -= 5
print (x)

x *= 5
print(x)

x /= 5
print (x)

x **= 5
print (x)

x //= 5
print (x)

# 5.Bitwise operator

# These operators perform bitwise operations on binary value of numbers.
# AND, OR, XOR, NOT, Left Shift and Right Shift come under Bitwise Operator.
# Ex:

a = 5  # Binary: 101
b = 3  # Binary: 011
print (a & b)  # Output: 1 (Binary: 001)
print (a | b)  # Output: 7 (Binary: 111)
print (a ^ b)
print (5 << 1)
print (5 >> 1)

# 6.Identity operator

# These operators are used to compare objects to see if they share the same memory location.
# Returns boolean value (True or False) as output.
# Ex:

x = [1, 2, 3]
y = x
z = [1, 2, 3]
print(x is y)
print(x is z)

# 7.Membership operator

# These operators test if a value or variable is present in a sequence (like strings, lists, tuples, etc).
# Ex:

my_list = [1, 2, 3, 4]
print(3 in my_list)      # Output: True
print(5 not in my_list)  # Output: True

# 8.Conditional operator

# This gives output as per conditional expressions.
# Ex:

a = 10
b = 20
min_val = a if a < b else b
print(min_val)  # Output: 10



# By Understanding function of each operator, helps us to write more efficient and readable code.
```

```python
# 5. Explain the concept of type casting in Python with examples.

#Ans:

# Type casting in Python refers to the process of converting one data type into another. 
# This is useful when we perform operations / computation there can be mismatch between the data type of the given data or there might be a requirement a specific data type.
# It ensures compatibility and preventing errors during execution of code.
# Python provides several built-in functions for casting between different types like str(), int(), list(), tuple() etc.
#There are 2 types of Casting in python;

# 1.Implicit Type Casting:

#Python automatically converts one data type to another while performing any operation whenever required.
#Ex:
# Implicit type casting

x = 5   # Integer
y = 3.5 # Float

result = x + y  # Python converts x to float before adding
print(result)   # Output: 8.5
print(type(result))  # Output: <class 'float'>


# 2.Explicit Type Casting:

# This requires the programmer to manually convert the data type using required built-in functions like int(), float(), str(), list(), tuple() etc.

# Ex:1 (Converting Float to Integer)
# Explicit type casting
a = 10.6
b = int(a)  # Casting float to integer
print(b)  # Output: 10 (Note: the decimal part is removed)

# Ex:2 (Converting String to Integer)
# Explicit type casting
num_str = "100"
num_int = int(num_str)  # Convert string to integer
print(num_int)  # Output: 100
print(type(num_int))  # Output: <class 'int'>

# Ex:3 (Converting Integer to String)
# Explicit type casting
x = 50
x_str = str(x)  # Convert integer to string
print(x_str)  # Output: '50'
print(type(x_str))  # Output: <class 'str'>

# Ex:4 (List to Tuple and Vice Versa)
# Converting list to tuple
my_list = [1, 2, 3]
my_tuple = tuple(my_list)
print(my_tuple)  # Output: (1, 2, 3)

# Converting tuple to list
new_list = list(my_tuple)
print(new_list)  # Output: [1, 2, 3]

```

```python
# 6. How do conditional statements work in Python? Illustrate using examples.

# Ans:

# Conditional statements in Python is a type of flow control mechanism.
# It decides the the execution of specific blocks of code based on whether a condition is True or False.
# These are used for decision-making in a program.

#Types of Conditional Statements in Python:

# 1.if Statement

#The if statement executes a block of code only if the given condition is True. Otherwise no output is returned.

# Ex:1
x = 10
if x > 5:
    print("x is greater than 5")  # This will be printed since the condition is True

# Ex:2
x = 3
if x > 5:
    print("x is greater than 5") # no output will be returned

# 2.if-else Statement

# The if-else statement allows us to execute one block of code if the condition is True and the 'else' block if the condition is False.

# Ex:1
x = 8
if x % 2 == 0:
    print(x, "is an even number") # This will be printed because the condition is true
else:
    print(x, "is an odd number")

# Ex:2

x = 5
if x % 2 == 0:
    print(x, "is an even number") 
else:
    print(x, "is an odd number") # This will be printed because the condition is false

 
# 3.if-elif-else statement

# The if-elif-else statement is used when there are multiple conditions to check. If the first condition is False, it checks the next condition using elif. If all conditions fail, the else block is executed as the final output.

# Ex:1
x = 10
if x < 10:
    print("x is less than 10")
elif x == 10:
    print("x is equal to 10")  # This will be printed
else:
    print("x is greater than 10")

#Ex:2
x = 18
if x < 10:
    print("x is less than 10")
elif x == 10:
    print("x is equal to 10")  
else:
    print("x is greater than 10") # This will be printed


# 4.Nested if statements

# If statement inside another if block is called as nested if statement.

# Ex: 
x = 8
y = 8
if x = 8:
        if y == 8:
        	print("x and y are equal")  # This will be printed since both conditions are true
else: 
	print ("X is different than y")
```

```python
# 7. Describe the different types of loops in python and their use cases with examples.

# Ans:

# Loops in Python is a type of flow control mechanism.
# Python provides two main types of loop: for loop and while loop. Both allow to repeatedly execute a block of code as long as a condition is met, but they have different use cases.

# 1. for Loop

# A for loop is typically used when the no. of iteration is known over a sequence (like a list, tuple, string, etc).

# Ex:1 (Looping through a list)

fruits = ["apple", "banana", "cherry"]
for item in fruits:
    print(item)

# Ex:2 (Using range() in a for loop)

for i in range(1, 6):  # Loops from 1 to 5
    print(i)

# Use Case:
# Iterating over elements in a sequence (list, tuple, string).
# Performing operations a fixed number of times (using range()).

# 2. while Loop

# A while loop repeats a block of code as long as a given condition is True. It is useful when the number of iterations is not known in advance.

# Ex:1
count = 1
while count <= 5:
    print(count)
    count += 1  # Incrementing count till count = 5

# Ex:2: (Infinite loop with a break statement)

while True:
    name = input("Enter your name: ")
    if name == "exit":
        break  # Exit the loop when 'exit' is typed
    print(f"Hello, {name}!")

# Use Case:

# Iterating when the number of iterations is not predetermined (e.g., waiting for user input, reading from a file until the end).
# Processing data while a condition remains true (e.g., a game loop running until the game is over).

#Control Statements in Loops:


# i)break Statement: Exits the loop when a certain condition is met.

# Ex:

for i in range(10):
    if i == 5:
        break
    print(i)

# ii)continue Statement: Skips the rest of the current loop iteration and moves to the next iteration.

# Ex:

for i in range(5):
    if i == 2:
        continue
    print(i)


# Nested Loops

# You can nest loops inside one another. For example, a for loop inside another for loop.

# Example: Nested for loop

for i in range(1, 4):
    for j in range(1, 4):
        print(f"i = {i}, j = {j}")
```
